```python
import os
import numpy as np
import json
import re
```

```python
def QMeasure(cluster_dict):
    """
    质量度量，评价聚类质量
    cluster: 分类后形成的簇群，一个字典，键为簇名，值为该簇的点
    """
    Measure = 0
    for key in cluster_dict.keys():
        cluster_points = cluster_dict[key]  # 形式为二维列表
        cluster_len = len(cluster_points)
        print('有%d个轨迹点' % cluster_len)
        cluster_measure = 0
        for point_i in cluster_points:
            for point_j in cluster_points:
                dist = np.sqrt(((np.array(point_i) - np.array(point_j))**2).sum() + np.spacing(1))
                cluster_measure += dist
        cluster_measure = cluster_measure/(2*cluster_len)
        print('第{0}类的度量为:{1}'.format(key, cluster_measure))
        Measure += cluster_measure
    return Measure
```

```python
nres = os.listdir('res')
days = [re.findall('\d{4}\d{2}\d{2}', nres[i])[0] for i in range(len(nres))]
for i in range(len(nres)):
    content = json.load(open('res/' + nres[i], encoding='gbk'))
    Measure = QMeasure(content)
    with open('value.txt', 'a', encoding='gbk') as f:
        f.write(days[i] + ' QMeasure:' + str(Measure) + '\n')
```
