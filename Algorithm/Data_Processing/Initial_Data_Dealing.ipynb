```python
import pandas as pd
import numpy as np
```

```python
#分块读取.csv大文件
reader = pd.read_csv('day20170224.csv', iterator=True)
#每个表都没有达到1亿的数据量,这里为了方便直接填了1亿
try:
    data = reader.get_chunk(100000000)
except Exception as e:
    print(e)
```

```python
data
```

```python
#剔除车辆状态为熄火的数据
data_ = data.drop(data[data.CAR_STAT1 == 7].index)
data_
```

```python
#为了后续加快运行速度,在这里直接删除了车辆状态这一列
data_ = data_.drop(['CAR_STAT1'], axis = 1)
data_
```

```python
#将Dataframe转化为列表
data_ = data_.values.tolist()
```

```python
#将每辆车与其对应的轨迹提取出来
from tqdm import trange
taxi = {}
for i in trange(len(data_)):
    temp = (data_[i][1], data_[i][2])
    if data_[i][0] not in taxi.keys():
        taxi[data_[i][0]] = []
    taxi[data_[i][0]].append(temp)
```

```python
#统计每辆车的轨迹点数
num = {}
for key in taxi.keys():
    num[key] = len(taxi[key])
#对轨迹点数进行排序
num = sorted(num.items(), key=lambda x:x[1], reverse=False)
num
```

```python
#删除前1500条和后2000条数据
del num[0:1500]
del num[-2000:-1]
num
```

```python
#随机抽取1k辆车
import random
sample = random.sample(num, 1000)
```

```python
#将所抽取的样本车辆及其对应的轨迹点提取出来
taxi_ = {}
for key in taxi.keys():
    for item in sample:
        if item[0] == key:
            taxi_[key] = taxi[key]
```

```python
#由于数据库中数据已经按入库时间排序好,根据观察,大概是每15秒进一次数据库
#我们将数据削减至原来的1/4,因此每辆车的轨迹点中,每四个条中删除前三条数据
_data = {}
#对样本车辆进行遍历
for key in taxi_.keys():
    temp = pd.DataFrame(taxi_[key], columns = ['lon', 'lat'])
    lenth = len(temp) - 4
    i = 0
    while i < lenth:
        temp = temp.drop([i,i+1,i+2])
        i += 4
    _data[key] = temp.values.tolist()
```

```python
#写入json文件贮存
import json
day_20170224 = json.dumps(_data, ensure_ascii = False)
fp1 = open('day_20170224.json', 'w+')
fp1.write(day_20170224)
fp1.close()
```
